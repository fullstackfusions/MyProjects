```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, title):
    # Generate a range of values for n
    n = np.linspace(1, 20, 1600)
    # Apply the provided function to each value in the range
    y = func(n)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(n, y, label=title)
    plt.title(title)
    plt.xlabel('n')
    plt.ylabel('f(n)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
# Plotting n
plot_function(lambda n: n, 'Linear - f(n) = n')

# Plotting log(n)
plot_function(lambda n: np.log(n), 'Logarithmic - f(n) = log(n)')

# Plotting n^1/3
plot_function(lambda n: n**(1/3), 'Quadratic - f(n) = n^1/3')

# Plotting n^1/2
plot_function(lambda n: n**(1/2), 'Quadratic - f(n) = n^1/2')

# Plotting n^2
plot_function(lambda n: n**2, 'Quadratic - f(n) = n^2')

# Plotting n^3
plot_function(lambda n: n**3, 'Quadratic - f(n) = n^3')

# Plotting n^100
plot_function(lambda n: n**100, 'Quadratic - f(n) = n^100')

# Plotting 2^n
plot_function(lambda n: 100**n, 'Quadratic - f(n) = 2^n')
```
